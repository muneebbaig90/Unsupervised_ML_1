# Importing  Libraries

```python
## import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataset = pd.read_csv("Market_Basket_Optimisation.csv", header=None)
transactions = []
## apriori requires a str lists of list
for i in range(0, 7501):
  transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
```

# Training

```python
## Training the apriori model on the dataset
from apyori import apriori
## min_support means which product is being sold everyday and how many times if the number is 3 or more then we use that number and multiply it with a week since the 7000+ data of tranactions is based on a week transactions
## (3 x 7)/7000+ some data
## confidence : u can say it generates rules meaning relation among data..example someone buys a burger that person will also buy french fries
## min_lift measures the quality of the rules
## min_length and max_length is the condition that defines how many elements u need to compare and make rules on
rules = apriori(transactions=transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

# Visualization

```python
## Visualizing the results
results = list(rules)
results
## support 0.0045 means that this is the percent that the these products are in how many of the 7000+ transactions
```

# Readable Format

```python
## more readable data
def inspect(results):
  product1 = [tuple(result[2][0][0])[0] for result in results]
  product2 = [tuple(result[2][0][1])[0] for result in results]
  supports = [result[1] for result in results]
  confidence = [result[2][0][2] for result in results]
  lift = [result[2][0][3] for result in results]
  return list(zip(product1, product2, confidence, lift))
resultsinDataframe = pd.DataFrame(inspect(results), columns = ['Product 1', 'Product 2', 'Confidence', 'Lift'])
```
